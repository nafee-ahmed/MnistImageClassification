# Importing libraries
The required libraries have been imported below:

```python
import numpy as np

import torch # using pytorch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
```

# Preparing the data
In order to prepare the data, several steps are done, and they are:

1. Reading the data: The data has been loaded into memory from the dataset
2. Separating data and labels: The first column contains the label, so it is separated to the variable called labels, and the remaining columns remain on the variable, data, as it is feature variables. The feature variables contain pixel information to construct the images.
3. Normalizing the data: Then normalization is done on the feature variables. This is done because if different feature variables contain different ranges, the deep learning model develops a bias towards feature variables with larger magnitudes. Normalizing the data in neural networks aims to bring the input data into a consistent range or distribution.
4. Splitting the data: Here, the splitting is done to separate the training and validation sets. The training set is used to update the model's weights, as the model compares its output with the actual output. Difference between the model's output and the actual output, causes the model to update the weights that brings its output closer to the actual output. 
    
    On the other hand, the validation set plays a role in both for training and for evaluation. It is a separate dataset that gets used during the training process to evaluate the model's performance and aid in hyperparameter tuning in every iteration of the training process. Since the model gets trained on the training set and hyperparameters trained on the training set, validation set gives an idea how the model performs on new data on those particular hyperparameter settings.

```python
# reading the data
data = np.loadtxt(open('mnist/mnist_train_small.csv', 'rb'), delimiter=',')
# Result below means:
# 20,000 = no. of images
# 785 = images of size 28x28=784, extra 1 = contains labels
data.shape
```

The cell below selects the samples i.e the feature variables and the corresponding labels

```python
# [x,y] for selection of part of matrix which means x is for selection of rows, and y is for selection of columns
labels = data[:,0] # means select all rows and only the first column. 
data = data[:,1:]  # means select all rows, and select from first column
print(f"Feature variables shape: {data.shape}") # (20000, 784) means 20,000 rows, and 784 columns
print(f"Target variables shape: {labels.shape}") # (20000,) means 20,000 rows, and 1 column
```

The cell below explores how to output the image using the pixel information on the feature variables from the csv dataset

```python
# prepare a plot of 3 rows and 4 columns with a size of (10,6)
fig, axs = plt.subplots(3,4,figsize=(10,6))
# this line loops over all the subplots in the 3x4 grid, flattening
# the array of subplots into a 1 dimensional array
for ax in axs.flatten():
    # generates a random integer between 0 and the number of images in the dataset
    rand_img_show = np.random.randint(0, high=data.shape[0])
    # reshapes the randomly selected image from a 1-dimensional array to a
    # 2-dimensional array of size 28x28
    img = np.reshape(data[rand_img_show,:], (28,28))
    ax.imshow(img, cmap='gray') # show the image
    ax.set(title=f'The number {labels[rand_img_show]}') # set the title
plt.suptitle('How we see the data') # set the suptitle
# adjust the spacing between the subplots
plt.tight_layout(rect=[0,0,1,0.95])
plt.show() # show the 3x4 plot
```

Here, the data is being normalized. As explained above, it is done so all the columns have consistent ranges so that the model does not develop a bias.

```python
data_norm = data / np.max(data) # every item in data array is divided by max value in array
```

On the cell below, the function split_data splits the data into validation and training tests: 
1. The line `data_t = torch.tensor(data_norm).float()` and `labels_t = torch.tensor(labels).long()` convert the feature variables and target variables into tensors so that pytorch can work with them because pytorch prefers its own format. The `.long()` and `.float()` are there to meet the format of pytorch as pytorch wants the feature variables and target variables to be in float and long respectively.
2. Then the line `train_data, test_data, train_labels, test_labels = train_test_split(data_t, labels_t, test_size=0.1)` splits the tensors to training and validation sets using scikit learn's method. In this case, the training and test set are split into the portion of 90%-10%.
3. The lines `train_data = TensorDataset(train_data, train_labels)` and `test_data = TensorDataset(test_data, test_labels)` converted to TensorDataset, which is a class of Pytorch for handling datasets. In this TensorDataset format, the data and the corresponding labels are combined to a single dataset object. 
4. The line `batch_size = 32`, means there are 32 samples in each batch. Weights are updated after each batch instead of updating after every single sample. The more detailed explanation of batch training is given when it gets to the training part.
5. The lines `train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True, drop_last=True)` and `test_loader = DataLoader(test_data, batch_size=test_data.tensors[0].shape[0])` convert to DataLoader object for efficient loading and batching of data during training. It also makes sure that the format is compatible with Pytorch so pytorch can work with it.

```python
def split_data(data_norm, labels):
    data_t = torch.tensor(data_norm).float() # convert normalized feature variables to tensor
    # convert target variable to tensor. .long() added so it is compatible with fomat that pytorch requires
    labels_t = torch.tensor(labels).long()
    # splits training and validation sets in portion of 90%-10%
    train_data, test_data, train_labels, test_labels = train_test_split(data_t, labels_t, test_size=0.1)
    # convert training set to TensorDataset, so labels and data are bundled together for pytorch
    train_data = TensorDataset(train_data, train_labels)
    # convert validation set to TensorDataset, so labels and data are bundled together for pytorch
    test_data = TensorDataset(test_data, test_labels)
    batch_size = 32 # for batch training. Explained in detail later on during training explanation
    # convert training set to DataLoader for pytorch compatibility
    # sets the batch size for batch training and shuffles the data
    # each sample must have equal numbers during training to reduce bias. 
    # that's why last sample is dropped in case it ran out of data to maintain equal batch sizing
    train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True, drop_last=True)
    # convert validation set to DataLoader object for pytorch compatibility
    # does not set the batch size as it equals to length of number of samples. batching is needed for training.
    test_loader = DataLoader(test_data, batch_size=test_data.tensors[0].shape[0])
    # return the results
    return train_loader, test_loader
train_loader, test_loader = split_data(data_norm, labels) # calls the function to split
```

# Creating the neural network
The cell below defines the neural network architecture for image classification to identify out of 10 images (digits 0 to 9) from the mnist dataset. The neural network consists of the input layer, several hidden layers, and the output layer. Each layer has a specified number of neurons that process the data, for correct predictions.

The class takes in the number of neurons for each hidden layer, and the number of hidden layers with those set neurons as parameters to construct the neural network architecture. Since these are taken as parameters, different models can be made with different layers and experimented on. 

Explanations for the reasoning of the layers are explained below:
1. Input layer: It is the first layer in the neural network. It receives the input as passes onto the next hidden layer for further processing. In this case, the input layer has been set to have 784 neurons since there are 784 feature variables, that construct the image.
2. Hidden layers: They are the layers in between the input layer and the output layer. The number of hidden layers and the number of neurons to be in each hidden layer are passed in as parameters, so they can be altered to understand which model is the best.
3. Output layer: It is the final layer and is responsible for the prediction based on the processing from the previous layers. The number of neurons in this layer is based on the problem. For example, in a classification task with multiple classes, the output layer may have one node per class, where each node represents the probability of the input belonging to that class. The node with the highest probability is often considered the predicted class label. In this case, the number of neurons is 10, since there are 10 classes (digits 0 to 9), and the highest probability of a class is chosen as the prediction.

**Choice of activation functions?**

On the `forward()` method, it can be seen `F.relu` being applied onto the hidden layers. It is an activation function that is being applied. It is applied to hidden layers to introduce non-linearity so that the model can learn complex patterns in the data. Relu is commonly chosen for hidden layers.

```python
# class defining neural network architecture
class ann_mnist(nn.Module):
    # gets called when class is initialized
    def __init__(self, n_neurons, n_layers):
        # initializing the layers
        super().__init__()
        # create dictionary to store the layers
        self.layers = nn.ModuleDict()
        self.n_layers = n_layers # stored as attribute of the class, so it is available throughout the class
        # input layer, 784 implying number of input features
        self.layers['input'] = nn.Linear(784, n_neurons)
        # number of hidden layers based on parameter passed 
        for i in range(n_layers):
            self.layers[f'hidden{i}'] = nn.Linear(n_neurons, n_neurons)
        # output layer
        self.layers['output'] = nn.Linear(n_neurons, 10)
    # forward prop
    def forward(self, x):
        # initializing the operations on the layers
        x = self.layers['input'](x) # passed to input layer
        # hidden layers
        for i in range(self.n_layers):
            x = F.relu(self.layers[f'hidden{i}'](x)) # applying relu activation function on hidden layer
        x = self.layers['output'](x) # passed to output layer
        return x # return prediction
```

# Training several models

In order to understand, how the neural network is getting trained, the basics have to be explained:

**How does neural network work?**

In each neuron, an input(s) is taken and an output is given, which can then serve as an input to another neuron, and that neuron gives an output. This way it is passed to different neurons, until finally a neuron gives the final result. That is how results are being passed from input layer to hidden layers to output layer.

**How does neural network train itself?**

Each neuron in the neural network performs a version of this, which is either linear regression or logistic regression such as if it is linear regression, each neuron does this:

Linear regression uses y = mx + c equation to make predictions.

Model: $f_{w,b} = wx + b$

Parameters: w, b

Cost Function: $J(w, b) = \frac{1}{2m}\sum_{i=1}^m (f_{w, b}(x^i) - y^i)^2$

The neuron receives inputs from the previous layer, which is the x, and it compares the actual value with the predicted value using the cost function (J), and lower the difference between the predicted and actual values, the lower the cost function (J). The output of the neuron serves as input to the neuron on next layer such as output of neuron from layer 1 serves as input of neuron from layer 2. This is known as forward propagation

Objective: Lesser the J, the better. Here, $f_{w, b}(x^i)$ is the predicted y and $y^i$ is the actual y and then multiple differences are calculated from predicted y and actual y. The parameters yielding the lowest J, is taken as the equation to make predictions. `m` is the number of rows.

With each sample the neural network takes, the model calculates the cost function. The model then updates it weights, w and b, so the cost function is lower the next time. And as the model learns, the cost functions keeps on getting lower. The training continues until the cost function does not change significantly or until the model has run out of iterations.

Here, forward propagation is the process of passing the data from input layer to hidden layers and eventually to the output layer. Every neuron performs the calculation just described aboved. When the data gets to the output layer, the data is then passed in the opposite direction towards the input layer, using a process known as backward propagation. The goal of backward propagation is to adjust the weights based on negative feedback. 

**Why batch training?**

Batch size is often in powers of 2.

Batching is a form of regularization. It smoots learning by averaging the loss over many samples, and thereby reduces overfitting. 

When we observe the data is too similar to each other, then the smaller the batch, it often gives better results. 

<img src='./Images/why batch training.png' width='440px' heigh='440px' alt='Batch Training advantages' />

On the previous section on how the model trains itself, it has been described that the model updates its weights after taking in every individual sample. But, this process of calculating the cost function and updating weights with each sample, the model can go down a longer path of minima, where the cost function is lowest. 

This is why batch training is done. With batch training, instead of updating the weights after every individual sample, the model updates its weights after processing a batch of examples. This gives an impression of taking an averaged out reading, so the reading is more precise. This reduces the frequency of weight updates and can lead to faster convergence during training.

**Why Adams algorithm instead of gradient descent in this code?**

First, what is learning rate need to understood. Learning rate is used during the training process. Learning rate is a hyperparameter in the training of neural networks. It determines the step size at which the model's parameters are updated during the training process.

Let Alpha be learning rate

If this was to start gradient descent, one step of gradient descent, and if Alpha is small, it may take a little bit in that direction, where the minima for cost function is. Then another step, then another step, then another step, then another step, and it is noticed that every single step of gradient descent is pretty much going in the same direction, and if this is to be the case, it might be wondered, well, why is it not the Alpha bigger, can the algorithm automatically increase Alpha? The same applies when the learning rate is too large, it has to be iterated and iterated, taking too big of a step every time and never reaching the minima.

It would be better, if the algorithm adjusted the learning rate by itself, such as if it sees, it is going in the same direction taking too small a step, then it should make the learning rate bigger, to help get to minima faster, and the same applies when the learning rate is too large. This is where Adam algorithm comes in.

This is what the code below translates to, now the comments on the code cell below dictates where each of the concepts are being applied

```python
# this function defines the model and all the variables needed for the model
def create_model(n_units_per_layer, n_layers):
    net = ann_mnist(n_units_per_layer, n_layers) # initializes the model
    # cost function is the average of loss function across all the samples. 
    # CrossEntropyLoss since this is a multi-class classification i.e choose an image across many images.
    loss_func = nn.CrossEntropyLoss() 
    optimizer = torch.optim.Adam(net.parameters(), lr=0.01) # defines the Adam algorithm
    return net, loss_func, optimizer # returns the results
# this function defines the training process and the evaluation of the models
def train_model(net, loss_func, optimizer, train_loader, test_loader):
    num_epochs = 60 # iterations set to 60
    losses = torch.zeros(num_epochs) # defines a tensor array of 0s to store losses
    train_acc = [] # array to record accuracy on training set after each epoch
    test_acc = [] # array to record accuracy on validation set after each epoch
    for epochi in range(num_epochs): # repeats every epoch
        batch_acc = [] # array to store accuracy for each batch
        batch_loss = [] # array to store losses for each batch
        for x,y in train_loader: # iterate over each batch in the training data
            # forward prop
            y_hat = net(x)
            # loss function
            loss = loss_func(y_hat, y)
            # backward prop
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            batch_loss.append(loss.item()) # store the loss values to batch_loss
            
            # stores in accuracy of the batch on batch_acc
            # torch.argmax(y_hat,axis=1) == y means it takes the class(i.e digit 0-9) with the highest probability 
            # as the prediction and then compares with the actual value
            batch_acc.append(100*torch.mean( (torch.argmax(y_hat,axis=1) == y).float() ).item())
        # the training accuracy is the average of accuracies across all the batches in that epoch
        train_acc.append(np.mean(batch_acc))
        # the training loss is the average of losses across all the batches in that epoch
        losses[epochi] = np.mean(batch_loss)
        # training ends here from this line, now testing is done on the validation set to calculate test accuracy
        # for every epoch
        batch_test_acc = [] # array to store the test accuracies across each batch
        
        for x,y in test_loader: # iterate over the test loader
            with torch.no_grad(): # eliminates unnecessary computation for faster processing
                y_hat = net(x) # predict the values based on x
            # batch_test_acc is calculated based on the same procedure that training accuracy was calculated on
            batch_test_acc.append(100*torch.mean( (torch.argmax(y_hat, axis=1) == y).float() ).item())
        
        test_acc.append(np.mean(batch_test_acc)) # test accuracy is the mean of accuracies of batch accuracies
    return train_acc, test_acc, losses, net # return the results
```

```python
num_layers = np.arange(1, 11) # array that goes from 1 to 10, representing the number of hidden layers
train_accs = [] # array to store the average training accuracy for each number of layers
test_accs = []  # array to store the average test accuracy for each number of layers
final_losses = [] # array to store the final loss value for each number of layers
for n_layer in range(len(num_layers)): # iterate over each number of layers
    
    # create the neural network model, loss function, and optimizer
    # the number of neurons in each hidden layer is set at a constant of 50
    # the number of hidden layers getting experimented on
    net, loss_func, optimizer = create_model(50, num_layers[n_layer])
    
    # train the model and obtain the training accuracy, test accuracy, losses, and the trained model
    train_acc, test_acc, losses, net = train_model(net, loss_func, optimizer, 
                                              train_loader, test_loader)
    
    train_accs.append(np.mean(train_acc[-5:])) # store the average training accuracy for the last 5 epochs
    test_accs.append(np.mean(test_acc[-5:])) # store the average test accuracy for the last 5 epochs
    final_losses.append(losses[-1]) # store the final loss value of the last epoch
```

# Plotting to pick the best model

Out of the 10 models experimented with different hidden layers, only 1 model is chosen as the best model to make predictions.

**How to choose the best model?**

The key is choose the model with the highest accuracy and the lowest loss. 

Accuracy refers to the percentage of correctly classified samples out of the total samples. It measures how well the model predicts the correct class labels. Higher accuracy indicates better performance.

Loss, on the other hand, is a measure of the dissimilarity between the predicted output of the model and the true labels. The lower the loss value, the closer the predicted outputs are to the true labels. Minimizing the loss during training helps improve the model's performance.

In this case, the best model is the one with 2 layers.

```python
fig, ax = plt.subplots(1,2,figsize=(10,5))
plt.subplots_adjust(hspace=0.5)  # Adjust the vertical spacing between subplots
ax[0].plot(num_layers, train_accs, 'o-', markerfacecolor='w', markersize=9, label='Train')
ax[0].plot(num_layers, test_accs, 'x-', markerfacecolor='r', markersize=9, label='Test')
ax[0].set(xlabel='Nuber of layers', ylabel='Accuracy', title='Accuracies')
ax[0].legend()

ax[1].plot(num_layers, final_losses, 'x-', markerfacecolor='w', markersize=9, label='Losses')
ax[1].set(xlabel='Number of layers', ylabel='Final Loss', title='Final Losses')

ax[0].set_xticks(range(int(min(num_layers)), int(max(num_layers)) + 1, 1))
ax[1].set_xticks(range(int(min(num_layers)), int(max(num_layers)) + 1, 1))
plt.show()
```
